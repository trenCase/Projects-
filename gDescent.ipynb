# Finding the minima using slope/derevative 
## as the slope-> 0 local minima is near 

```python
def dfunc(x) : 

    return(2*x + 1)
```

```python

plt.figure(figsize=[10,5])
plt.grid()
plt.title('Slope of Cost Function')
plt.ylabel('derivative of f(x)')
plt.xlabel('x')
plt.plot(x,dfunc(x))
plt.show()
```

## An example Gradient Descent 

```python
curr_x = 4 
temp_x = 0 
learning_rate = 0.1 

for n  in range(50) : 
    temp_x = curr_x 
    grad = dfunc(temp_x)
    curr_x = temp_x - grad * learning_rate 

print(f'local minima : {curr_x}')
print(f'value of slope  : {dfunc(curr_x)}')
```

# Multivariate Gradient Descent


$ g(x) = x^4 - 4x^2 + 5 $

```python
def gfunc(x) : 
    return x**4 - 4*x**2 + 5 
```

```python
def dgfunc(x) : 
    return 4*x**3 - 8*x 
```

```python
test_1 = np.linspace(-3,3,1000)

plt.figure(figsize=[10,5])
plt.xlim(-3,3)
plt.ylim(0,6)
plt.title('Cost function')

plt.ylabel('g(x)')
plt.plot(test_1,gfunc(test_1))
plt.show()
```

```python
plt.figure(figsize=[10,5])

plt.title('Slope of Cost function')
plt.ylim(-5,5)
plt.xlim(-2,2)
plt.grid()
plt.ylabel('derivative of g(x)')
plt.plot(test_1,dgfunc(test_1))
plt.show()
```

```python
def gradient_descent(d_func, initial_x , learning_rate , max_iter = 500 ) : 

    curr_x = initial_x 
    list_x = [curr_x]
    slope_list = [d_func(curr_x)]

    for i in range(max_iter) : 

        temp_x = curr_x 
        gradient = d_func(temp_x)
        curr_x = temp_x - learning_rate * gradient

        step_size = abs(curr_x - temp_x) 

        list_x.append(curr_x) 
        slope_list.append(d_func(curr_x)) 

    return (curr_x , list_x , slope_list)

```

```python
local_min,list_x , derivative_list = gradient_descent(dgfunc,-0.5,0.02)
```

# 3-D Charts 

## min. $$f(x,y) = \frac{1}{3^{-x^2 - y^2} + 1} $$

```python
def f(x,y) : 
    r = 3 ** (-x**2 - y**2)
    return 1/(r+1)
    
```

## Plotting and Creating Data 

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

dataX = np.linspace(-2, 2, 200)
dataY = np.linspace(-3,3,200)

dataX , dataY  = np.meshgrid(dataX, dataY)

fig = plt.figure(figsize = [15,12])
ax = fig.gca(projection= "3d")

ax.set_xlabel('dataX')
ax.set_ylabel('dataY')
ax.set_zlabel('Cost (f(x,y)')

ax.plot_surface(dataX, dataY, f(dataX,dataY), cmap = 'coolwarm', alpha = 0.5)

```

## the red areas indicate high cost and the blue areas indicate low costs 

# Partial Derivatives and Symbolic Computations

```python
f(a,b).evalf(subs = {a:1.8, b:1.0})
```

```python
dFx = diff(f(a,b),a)
dFy = diff(f(a,b), b)
```

# Batch Gradient Descent 

```python
multiplier =  0.1
maxIter = 200 
params = np.array([1.8, 1.0]) 
valArray = params.reshape(1,2)


for n in range(maxIter) : 
    gradientX = dFx.evalf(subs = {a : params[0], b:params[1]})
    gradientY = dFy.evalf(subs = {a : params[0], b:params[1]})
    gradients = np.array([gradientX, gradientY])
    params = params - multiplier * gradients 
    valArray = np.append(valArray, params.reshape(1,2), axis = 0 )

zValues = f(valArray[:, 0], valArray[:, 1])

print(gradients)
print(f(params[0], params[1]))
print(valArray.shape)
#print(zValues)
```

## Mean Squared Errors and Residual Square Errors

### $$RSS = \sum_{i=1}^n \big(y^{(i)} - h\theta(x)^{(i)}\big)^2 $$

### $$MSS = \frac{1}{n}\sum_{i=1}^n \big(y^{(i)} - h\theta(x)^{(i)}\big)^2 $$



```python
from sklearn.linear_model import LinearRegression 

dataX = np.array([0.1, 1, 1.5 , 2.4, 3.6, 5.1 , 6.7])
dataY = np.array([1.7, 2.8, 3.5 , 3.0 , 8.4 , 9.0 , 8.0])

regr = LinearRegression() 
regr.fit(dataX.reshape(7,1),dataY.reshape(7,1))

fig = plt.figure(figsize = [15,12])
ax = fig.add_subplot(1,1,1)

ax.scatter(dataX, dataY , s= 40 , color = 'blue')
ax.plot(dataX, regr.predict(dataX.reshape(7,1)), color = 'red' , linewidth = 3)
```

```python
rI = regr.intercept_[0]
rI 
```

```python
rC = regr.coef_[0][0]
rC 
```

```python
hx = rI + rC * dataX
hx = hx.reshape(7,1)
dataY = dataY.reshape(7,1)
dataX = dataX.reshape(7,1)
print(hx)
print(dataY)
```

```python
def mse(x , y) : 

    mse  = np.average((x-y)**2, axis = 0)
    return mse 

from sklearn.metrics import mean_squared_error

print(mse(dataY, hx))
print(mean_squared_error(dataY, regr.predict(dataX)))
```

```python
## Partial Derivative of MSE 

def grad4(x, y, thetas) : 
    n = y.size 
    z = sum(y - thetas[0] - thetas[1]*x)

    theta0Slope = (-2/n) * z 
    theta1Slope = (-2/n) * sum(z*x)

    return np.array([theta0Slope[0], theta1Slope[0]])
```

```python
multiplier = 0.01 
thetas = np.array([2.9,2.9])

plotVals = thetas.reshape(1,2)
mseVals = mse(dataY, thetas[0] + thetas[1]*dataX)

for i in range(1000) : 
    thetas = thetas - multiplier * grad4(dataX, dataY, thetas )

    plotVals = np.append(plotVals, thetas.reshape(1,2),axis = 0)
    mseVals = np.append(mseVals, mse(dataY, thetas[0] + thetas[1] * dataX))
    
    

print(thetas[0])
print(thetas[1])
print(mean_squared_error(dataY, regr.predict(dataX)))
print(mean_squared_error(dataY, thetas[0] + thetas[1]*dataX))
plotVals.shape
```

```python
fig = plt.figure(figsize= [ 15,12])
ax  = fig.gca(projection = '3d')

ax.set_xlabel("theta 0")
ax.set_ylabel("theta 1")
ax.set_zlabel("cost - mse")

ax.scatter(plotVals[:,0], plotVals[:, 1], mseVals, s=50, color = 'black')
ax.plot_surface(plot0, plot1, plotCost, cmap = 'coolwarm',alpha = 0.5)

plt.show()
```
